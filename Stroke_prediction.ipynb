```python
stroke_prediction_DS.loc[stroke_prediction_DS['ever_married'] == 'Yes', 'ever_married'] = 1
stroke_prediction_DS.loc[stroke_prediction_DS['ever_married'] == 'No', 'ever_married'] = 0
```

```python
stroke_prediction_DS.head()
```

```python
stroke_prediction_DS.loc[stroke_prediction_DS['gender'] == 'Female', 'gender'] = 1
stroke_prediction_DS.loc[stroke_prediction_DS['gender'] == 'Male', 'gender'] = 2
stroke_prediction_DS.loc[stroke_prediction_DS['gender'] == 'Other', 'gender'] = 3
```

```python
stroke_prediction_DS.head()

```

```python
def replace_work_type(x):
    if x=='children':
        return 0
    elif x=='Private':
        return 1
    elif x=='Govt_job':
        return 2
    elif x=='Never_worked':
        return 3
    elif x=='Self-employed':
        return 4


stroke_prediction_DS['work_type']=stroke_prediction_DS['work_type'].apply(replace_work_type)
```

```python
def replace_smoking_status(x):
    if x=='formerly smoked':
        return 0
    elif x=='never smoked':
        return 1
    elif x=='smokes':
        return 2
    elif x=='Unknown':
        return 3



stroke_prediction_DS['smoking_status']=stroke_prediction_DS['smoking_status'].apply(replace_smoking_status)
```

```python
stroke_prediction_DS.loc[stroke_prediction_DS['Residence_type'] == 'Urban', 'Residence_type'] = 1
stroke_prediction_DS.loc[stroke_prediction_DS['Residence_type'] == 'Rural', 'Residence_type'] = 2
```

```python
labels = stroke_prediction_DS['stroke']
(stroke_prediction_DS.smoking_status == 'Unknown').sum()
```

```python
labels = stroke_prediction_DS['stroke']

import copy
stroke_prediction_DS_backup = copy.deepcopy(stroke_prediction_DS)
```

```python
labels
```

```python
stroke_prediction_DS.drop('stroke',axis=1,inplace=True)
```

```python
stroke_prediction_DS
```

```python
from sklearn.model_selection import train_test_split

train_features, test_features = train_test_split(stroke_prediction_DS, test_size=0.3)
train_labels, test_labels = train_test_split(labels, test_size=0.3)
```

```python
train_features
```

```python
train_features
```

```python
#Train model
from sklearn.ensemble import RandomForestClassifier
from sklearn.inspection import permutation_importance
from sklearn.metrics import accuracy_score, precision_score, recall_score,confusion_matrix
```

```python
rf_classifier = RandomForestClassifier(verbose=2,oob_score=True)
```

```python
rf_classifier.fit(train_features,train_labels)
```

```python
predictions = rf_classifier.predict(test_features)
```

```python
accuracy = 100.0 * accuracy_score(test_labels,predictions)
```

```python
accuracy
```

```python
feature_importances = rf_classifier.feature_importances_
```

```python
feature_importances
```

```python
rf_classifier.oob_score_
```

```python
import numpy as np
```

```python
std = np.std([tree.feature_importances_ for tree in rf_classifier.estimators_], axis=0)
indices = np.argsort(feature_importances)[::-1]
```

```python
features_plot = ['gender', 'age' ,'hypertension', 'heart_disease', 'ever_married' ,'work_type' ,'Residence_type' ,'avg_glucose_level' ,'bmi', 'smoking_status']
```

```python
indices_list = indices.tolist()
for label in range(len(indices_list)):
    indices_list[label] = features_plot[indices_list[label]]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.title('Rendesia e vetive ne marrjen e vendimit')
plt.bar(range(train_features.shape[1]), feature_importances[indices], color='r', yerr=std[indices], align = 'center')
plt.xticks(range(train_features.shape[1]), indices)
plt.xlim([-1,train_features.shape[1]])
plt.legend()
plt.show()
```

```python
results = permutation_importance(rf_classifier, test_features,test_labels,n_repeats=30, random_state=0, n_jobs=2)
```

```python
results.importances_mean
```

```python
sorted_idx = results.importances_mean.argsort()
```

```python
sorted_idx
```

```python
results.importances[sorted_idx].T
```

```python
fig, ax = plt.subplots()
ax.boxplot(results.importances[sorted_idx].T, vert=False, labels = sorted_idx)
ax.set_title("Permutation importances")
fig.tight_layout()
plt.show()
```
